```python
import matplotlib.pyplot as plt
from datetime import datetime
import numpy as np
import pandas as pd
```

```python
print("Number of packets in test1.csv - {}".format(pd.read_csv("data/test1.csv").shape[0]))
print("Number of packets in test2.csv - {}".format(pd.read_csv("data/test2.csv").shape[0]))
print("Number of packets in test3.csv - {}".format(pd.read_csv("data/test3.csv").shape[0]))
print("Number of packets in test4.csv - {}".format(pd.read_csv("data/test4.csv").shape[0]))
```

```python
def plot_using_csv(filename):
    # read csv file
    df = pd.read_csv(filename)

    # filter only incoming packets
    df = df.loc[df['Destination'] == '10.0.2.15']
#     print(df.shape)

    # filter only relevant columns
    df = df[['Time', 'Length']]

#     # print the 5 biggest packets
#     df_sorted = df.sort_values("Length", ascending=False)
#     print(df_sorted.head())

    plt.hist(df.Time, weights=df.Length, bins=df.shape[0])
    plt.xlabel("Time(Seconds)")
    plt.ylabel("Packet Length")
    plt.show()

    # get pdf
    time = list(df["Time"])
    inter_message_delays = [int(time[i+1] - time[i]) for i in range(len(time)-1)]

    plt.hist(inter_message_delays)
    plt.xlabel('Inter-Message Delay (Seconds)')
    plt.ylabel('PDF')
    plt.show()
```

# Graphs

```python
plot_using_csv("data/test1.csv")
```

```python
plot_using_csv("data/test2.csv")
```

```python
plot_using_csv("data/test3.csv")
```

```python
plot_using_csv("data/test4.csv")
```
